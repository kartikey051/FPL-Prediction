```python
driver.get("https://fbref.com/en/comps/9/wages/Premier-League-Wages") 
```

```python
# Locate the table row (example: first row of the table)
row = driver.find_element(By.XPATH, "//table/tbody/tr[1]")
```

```python
# Get all <td> elements (columns) inside that row
cols = row.find_elements(By.TAG_NAME, "td")
```

```python
# Extract text from each column and store in a list
col_values = [col.text for col in cols]
```
