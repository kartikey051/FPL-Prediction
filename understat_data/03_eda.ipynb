```python
plt.figure(figsize=(12, 6))
sns.histplot(match_df["home_goals"], bins=6, kde=True)
plt.title("Distribution of Home Goals")
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(match_df["away_goals"], bins=6, kde=True)
plt.title("Distribution of Away Goals")
plt.xlabel("Goals")
plt.ylabel("Frequency")
plt.show()
```

Both observations, the goal distribution is right-skewed.
The low-scoring games dominate.
The regression models must handle the imbalance

# xG vs Actual Goals

```python
sns.scatterplot(
    data=match_df,
    x="home_xg",
    y="home_goals",
    alpha=0.6
)
plt.title("Home xG vs Home Goals")
plt.xlabel("Expected Goals xG")
plt.ylabel("Actual Goals")
plt.show()
```

```python
sns.scatterplot(
    data=match_df,
    x="away_xg",
    y="away_goals",
    alpha=0.6
)
plt.title("Away xG vs Away Goals")
plt.xlabel("Expected Goals xG")
plt.ylabel("Actual Goals")
plt.show()
```

xG correlates but does not perfectly predict goals
xG is better for player quality than raw goals

# Goal Difference vs xG Difference

```python
sns.scatterplot(
    data=match_df,
    x="xg_diff",
    y="goal_diff",
    alpha=0.6
)
plt.axhline(0, color='red', linestyle='--')
plt.axvline(0, color='red', linestyle='--')
plt.title("xG Difference vs Goal Difference")
plt.xlabel("xG Difference")
plt.ylabel("Goal Difference")
plt.show()
```

```python
Use of xG-based metrics in Best 11 selection is justified.
```

# Player / Roster EDA

```python
sns.histplot(roster_df["time"], bins=20)
plt.title("Time Played Distribution")
plt.xlabel("time")
plt.ylabel("Number of Players")
plt.show()
```

Most of the players have played 90 minutes. Those who have not played are excluded, as per-90 stats become unstable

## Goal vs xG (Over/Under Performers)

```python
sns.scatterplot(
    data=roster_df,
    x="xg",
    y="goals",
    alpha=0.6
)
plt.plot([0, roster_df["xg"].max()], [0, roster_df["xg"].max()], 'r--')
plt.xlabel("Expected Goals (xG)")
plt.ylabel("Goals")
plt.title("Goals vs Expected Goals (xG)")
plt.show()
```

Above the line lies the clinical finishers
Below the line lies the underperformers
